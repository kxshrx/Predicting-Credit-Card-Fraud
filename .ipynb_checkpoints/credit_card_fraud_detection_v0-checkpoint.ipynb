```python
xtrain, xtest, ytrain, ytest = train_test_split(df.iloc[:,:-1],df.iloc[:,-1],
                                               test_size = 0.3)
```

# BUILDING MODEL_V0

```python
model_v0 = XGBClassifier()
model_v0.fit(xtrain,ytrain)
ypred = model_v0.predict(xtest)
```

```python
print(f"Accuracy : {accuracy_score(ytest,ypred)*100}\n")
print(f"Confusion Matrix :\n {confusion_matrix(ytest,ypred)}")
```

```python
print(classification_report(ytest,ypred))
```

```python
sns.heatmap(confusion_matrix(ytest,ypred),annot=True,cmap="mako",linewidths = 0.7)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python

# Define models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC(probability=True),
    "k-NN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB()
}

# Train and evaluate models
for name, model in models.items():
    if name in ["Logistic Regression", "SVM", "k-NN"]:
        model.fit(sc.fit_transform(xtrain),ytrain)
        y_pred = model.predict(sc.fit_transform(xtest))
    else:
        model.fit(xtrain, ytrain)
        y_pred = model.predict(xtest)
    
    # Calculate accuracy
    accuracy = accuracy_score(ytest, ypred)
    
    # Print accuracy
    print(f"Accuracy for {name}: {accuracy:.4f}")
    
    # Print classification report
    print(f"Classification Report for {name}:")
    print(classification_report(ytest, ypred))
    print("---------------------------------------------")
    print("\n\n")
```
